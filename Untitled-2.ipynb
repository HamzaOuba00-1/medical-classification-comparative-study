## 2. Data Cleaning

This block prepares the dataset for reliable analysis and modeling.

## 5. Data Preprocessing

This section prepares the cleaned dataset for machine learning.  
The goal is to transform the data into a numerical, scaled, and balanced form suitable for predictive modeling.

The steps executed here follow best practices:

### 5.1 Use the cleaned dataframe  
We begin from the fully cleaned dataset (`df`), ensuring:
- no missing values,
- consistent categories,
- medically coherent records.

### 5.2 Encode categorical variables
We apply:
- **Label Encoding** for binary variables,  
- **One-Hot Encoding** for multi-category variables (`drop_first=True` to avoid collinearity).

This guarantees that all features become numerical.

### 5.3 Separate features and target
We construct:
- `X` = all predictors  
- `y` = stroke label  

### 5.4 Stratified Train–Test Split
Given the strong label imbalance, stratification ensures the same class proportions in train and test.

### 5.5 Standard Scaling
We normalize numerical features using **StandardScaler**, which stabilizes:
- logistic regression optimization,
- SVM kernel distances,
- distance-based oversampling (SMOTE).

### 5.6 SMOTE Oversampling (Train Only)
We apply **SMOTE only to the training data** to synthetically generate minority samples and obtain a 50/50 distribution.

This prevents:
- data leakage,
- contamination of the test set,
- unrealistic performance estimates.

After this block, we obtain:
- `X_train_smote`, `y_train_smote` → balanced training set  
- `X_test_scaled` → untouched, real-world distribution for fair evaluation  


```python
# ============================================================
# 5. Data preprocessing for modeling
#    - Encoding
#    - Train/test split
#    - Scaling
#    - SMOTE
# ============================================================

print("=== Starting preprocessing ===\n")

# Work on a fresh copy of the cleaned dataset
df_prep = df.copy()

# Remove useless identifier
if "id" in df_prep.columns:
    df_prep = df_prep.drop(columns=["id"])

# Binary categorical columns (will use LabelEncoder)
binary_cat_cols = ["gender", "ever_married", "Residence_type"]
# Multi-class categorical columns (will use one-hot encoding)
multi_cat_cols = ["work_type", "smoking_status"]

# Label encode binary categoricals
le = LabelEncoder()
for col in binary_cat_cols:
    df_prep[col] = le.fit_transform(df_prep[col])

# One-hot encode multi-class categoricals (drop_first to avoid full multicollinearity)
df_prep = pd.get_dummies(df_prep, columns=multi_cat_cols, drop_first=True)

print("Shape after categorical encoding:", df_prep.shape)

# Separate features and target
X = df_prep.drop(columns=["stroke"])
y = df_prep["stroke"]

print("\nX shape:", X.shape)
print("y shape:", y.shape)

print("\nMissing values in X (should be 0):", X.isna().sum().sum())

# --------------------------------
# 5.1 Train/Test split (stratified)
# --------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\n=== After train/test split ===")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("\ny_train class proportions:")
print(y_train.value_counts(normalize=True))
print("\ny_test class proportions:")
print(y_test.value_counts(normalize=True))

# --------------------------------
# 5.2 Scaling (StandardScaler)
#     (here we follow your pipeline: scale first, then SMOTE)
# --------------------------------
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

print("\nScaling done (StandardScaler).")
print("Any NaN in X_train_scaled? ->", np.isnan(X_train_scaled).any())
print("Any NaN in X_test_scaled?  ->", np.isnan(X_test_scaled).any())

# --------------------------------
# 5.3 SMOTE oversampling (on TRAIN only, in scaled space)
# --------------------------------
sm = SMOTE(random_state=42)
X_train_smote, y_train_smote = sm.fit_resample(X_train_scaled, y_train)

print("\n=== After SMOTE ===")
print("X_train_smote:", X_train_smote.shape)
print("y_train_smote class proportions:")
print(y_train_smote.value_counts(normalize=True))

print("\nPreprocessing COMPLETE: no missing values, balanced training set.")
```

# 6. Helper functions

```python
# ============================================================
# 6. Helper functions: evaluation and threshold calibration
# ============================================================

def evaluate_model(name, y_true, y_pred, y_proba, plot_roc=True, plot_pr=True):
    """
    Print standard classification metrics, show confusion matrix,
    and optionally ROC and Precision–Recall curves.
    """
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec  = recall_score(y_true, y_pred)
    f1   = f1_score(y_true, y_pred)
    auc_roc = roc_auc_score(y_true, y_proba)

    print(f"=== {name} — Test set performance (thr=0.50) ===")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")
    print(f"ROC-AUC  : {auc_roc:.4f}\n")

    cm = confusion_matrix(y_true, y_pred)
    print("Confusion matrix (rows = true, cols = predicted):")
    print(cm, "\n")

    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion matrix — {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    if plot_roc:
        fpr, tpr, _ = roc_curve(y_true, y_proba)
        plt.figure(figsize=(4, 3))
        plt.plot(fpr, tpr, label=f"AUC={auc_roc:.3f}")
        plt.plot([0, 1], [0, 1], "--", label="Random")
        plt.title(f"ROC curve — {name}")
        plt.xlabel("False positive rate")
        plt.ylabel("True positive rate")
        plt.legend()
        plt.tight_layout()
        plt.show()

    if plot_pr:
        precisions, recalls, _ = precision_recall_curve(y_true, y_proba)
        plt.figure(figsize=(4, 3))
        plt.plot(recalls, precisions)
        plt.title(f"Precision–Recall curve — {name}")
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.tight_layout()
        plt.show()


def calibrate_thresholds(name, y_true, y_proba, target_recall=0.90):
    """
    Compute and compare several thresholds:
    - default 0.50
    - F1-optimal
    - recall-optimal (>= target_recall)
    - Youden's J statistic based on ROC
    Prints metrics and confusion matrices for each,
    and plots the PR curve with the F1-optimal threshold.
    """
    precisions, recalls, thresholds_pr = precision_recall_curve(y_true, y_proba)
    thresholds_full = np.append(thresholds_pr, 1.0)

    # F1-optimal threshold
    f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-9)
    best_f1_idx = np.argmax(f1_scores)
    best_f1_thr = thresholds_full[best_f1_idx]

    # Recall-optimal threshold (>= target_recall)
    idx_recall = np.where(recalls >= target_recall)[0]
    best_rec_thr = thresholds_full[idx_recall[0]] if len(idx_recall) > 0 else thresholds_full[-1]

    # Youden's J threshold from ROC
    fpr, tpr, thresholds_roc = roc_curve(y_true, y_proba)
    youden_J = tpr - fpr
    best_J_idx = np.argmax(youden_J)
    best_J_thr = thresholds_roc[best_J_idx]

    thresholds = {
        "Default 0.50": 0.50,
        "F1-optimal": best_f1_thr,
        f"Recall-optimal ({target_recall:.0%})": best_rec_thr,
        "Youden J": best_J_thr
    }

    print(f"=== {name} — candidate thresholds ===")
    print(f"F1-optimal threshold             : {best_f1_thr:.4f}")
    print(f"Recall-optimal threshold ({target_recall:.0%}) : {best_rec_thr:.4f}")
    print(f"Youden J threshold               : {best_J_thr:.4f}\n")

    print("=== Evaluation of thresholds on TEST set ===\n")
    for label, thr in thresholds.items():
        y_pred_thr = (y_proba >= thr).astype(int)
        prec_thr = precision_score(y_true, y_pred_thr)
        rec_thr  = recall_score(y_true, y_pred_thr)
        f1_thr   = f1_score(y_true, y_pred_thr)
        cm_thr   = confusion_matrix(y_true, y_pred_thr)

        print(f"--- {label} (thr={thr:.4f}) ---")
        print(f"Precision: {prec_thr:.4f}")
        print(f"Recall   : {rec_thr:.4f}")
        print(f"F1-score : {f1_thr:.4f}")
        print("Confusion matrix:")
        print(cm_thr, "\n")

    # PR curve with F1-optimal threshold marked
    plt.figure(figsize=(6, 4))
    plt.plot(recalls, precisions, label="PR curve")
    plt.axvline(recalls[best_f1_idx], color="red", linestyle="--",
                label=f"F1-opt thr={best_f1_thr:.2f}")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title(f"Precision–Recall curve — {name}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return thresholds
```

# 7. Logistic Regression

```python
# ============================================================
# 7. Logistic Regression — CV, test evaluation, threshold tuning
# ============================================================

# Base model (no class_weight because SMOTE already balanced the training data)
log_reg_big = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    max_iter=2000,
    random_state=42
)

scoring = {
    "roc_auc": "roc_auc",
    "avg_precision": "average_precision",
    "recall": "recall",
    "f1": "f1",
    "accuracy": "accuracy"
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results_lr = cross_validate(
    log_reg_big,
    X_train_smote,
    y_train_smote,
    cv=cv,
    scoring=scoring,
    n_jobs=-1,
    return_train_score=False
)

print("=== Logistic Regression (SMOTE train) — 5-fold CV ===")
for metric in scoring.keys():
    scores = cv_results_lr[f"test_{metric}"]
    print(f"{metric:12s}: {scores.mean():.4f} ± {scores.std():.4f}")
print("\nROC-AUC per fold:", cv_results_lr["test_roc_auc"], "\n")

# Train final model on full SMOTE-balanced training set
log_reg_final_big = LogisticRegression(
    penalty="l2",
    C=1.0,
    solver="lbfgs",
    max_iter=2000,
    random_state=42
)

log_reg_final_big.fit(X_train_smote, y_train_smote)

# Probabilities and default predictions on the (imbalanced) test set
y_proba_lr_big = log_reg_final_big.predict_proba(X_test_scaled)[:, 1]
y_pred_lr_big  = (y_proba_lr_big >= 0.5).astype(int)

evaluate_model("Logistic Regression", y_test, y_pred_lr_big, y_proba_lr_big)

# Threshold calibration for logistic regression
lr_thresholds = calibrate_thresholds("Logistic Regression", y_test, y_proba_lr_big, target_recall=0.90)
```

## 7B. Logistic Regression — Test Set Evaluation (Threshold = 0.50)

We evaluate the trained model on the original, imbalanced test set.  
Using the default threshold of 0.50 leads to the typical trade-offs:

- **High recall** (detects most stroke cases)  
- **Very low precision** (many false positives)  
- Moderate ROC-AUC, but F1-score remains low due to imbalance  

The confusion matrix highlights the medical relevance:
- False negatives are rare (good),  
- False positives are numerous (expected in medical screening).  

Default thresholding is rarely optimal in healthcare; therefore we proceed with threshold calibration.

## 7C. Logistic Regression — Decision Threshold Calibration

The default classification threshold (0.50) is rarely optimal for imbalanced medical data.  
We therefore compute several clinically meaningful thresholds:

### 1. **F1-Optimal Threshold**  
Maximizes the harmonic mean between precision and recall.  
Useful when balancing false positives and false negatives.

### 2. **Recall-Optimal Threshold (≥90%)**  
Ensures high sensitivity — critical in healthcare screening, where missing a stroke-risk patient is costly.

### 3. **Youden’s J Statistic**  
Maximizes `(TPR − FPR)` and is widely used in diagnostic test evaluation.

Each threshold is evaluated on the test set, producing:
- updated precision/recall trade-offs,  
- different confusion matrices,  
- alternative operating points for clinical usage.  

These calibrated thresholds allow choosing the best strategy depending on the medical objective.

Interpretation:
- The curve illustrates the difficulty of the task due to severe imbalance.
- The F1-optimal point corresponds to a reasonable compromise between sensitivity and precision.
- Models with high AUC-PR are particularly valuable for rare events like stroke.

This plot also helps clinicians understand how performance changes as recall increases.

## 8D. Random Forest — Decision Threshold Calibration

Because the test set is extremely imbalanced, the default threshold (0.50)
is not meaningful. We therefore compute:

- **F1-optimal threshold**
- **Recall-optimal threshold (≥ 90%)**
- **Youden’s J threshold**

### Findings:
- **F1-optimal threshold = 0.16**  
  - recall increases sharply (0.70)  
  - but precision drops to 0.12  
  - F1 improves modestly (from 0.18 → 0.21)

- **Recall-optimal threshold (≥ 90%) = 0.00**  
  - all cases predicted positive  
  - perfect recall (1.00)  
  - unusable precision (0.05)

- **Youden’s J threshold = 0.1233**  
  - recall ≈ 0.78  
  - precision ≈ 0.11  
  - balanced improvement over the default threshold

### Interpretation:
Random Forest outputs **poorly calibrated probabilities**, forcing extremely low
thresholds to achieve acceptable recall. This behavior is common in models trained on
synthetic SMOTE minority samples.

Clinically, the model is unreliable for risk screening because it either:
- misses many cases (default threshold), or  
- produces excessive false positives (calibrated threshold).
